```python
from abc import ABC, abstractmethod
```

```python
class Car(ABC):
    
    @abstractmethod
    def maxSpeed(self):
        pass
```

```python
#myCar=Car()
```

```python
class Tesla(Car):
    def maxSpeed(self):
        print("200km")
```

```python
tesla=Tesla()
```

```python
tesla.maxSpeed()
```

```python
class Mercedes(Car):
    def maxSpeed(self):
     print("250km")
```

```python
mercedes = Mercedes()
```

```python
mercedes.maxSpeed()
```
